```python
sns.set_theme(style="whitegrid", palette="tab10")
plt.rcParams["figure.figsize"] = (10, 6)

df = pd.read_csv(path)
dfc = df.copy()
```

```python
# 1) Basic cleaning: ratings, cost, votes, city; remove duplicates
dfc['rate'] = pd.to_numeric(
    dfc['rate'].astype(str).str.extract(r'(\d+\.?\d*)', expand=False),
    errors='coerce'
)

if 'approx_cost(for two people)' in dfc.columns:
    dfc['approx_cost(for two people)'] = pd.to_numeric(
        dfc['approx_cost(for two people)'].astype(str).str.replace(',', ''),
        errors='coerce'
    )

if 'votes' in dfc.columns:
    dfc['votes'] = pd.to_numeric(dfc['votes'], errors='coerce')

dfc['city'] = dfc['listed_in(city)'] if 'listed_in(city)' in dfc.columns else np.nan

subset_cols = [c for c in ['name', 'address', 'location'] if c in dfc.columns]
if subset_cols:
    before = len(dfc)
    dfc = dfc.drop_duplicates(subset=subset_cols, keep='first')
    print(f"Removed {before - len(dfc)} duplicates based on {subset_cols}")
```

```python
# 2) Missing values overview
nulls = dfc.isna().sum().sort_values(ascending=False)
print("Top missing columns:\n", nulls.head(20))

numeric_cols = [c for c in ['rate', 'votes', 'approx_cost(for two people)'] if c in dfc.columns]
print("\nNumeric summary:\n", dfc[numeric_cols].describe().T)
```

```python
# 3) Univariate distributions
# Ratings
sns.histplot(data=dfc, x='rate', bins=20, kde=True)
plt.title('Distribution of Ratings')
plt.show()

sns.boxplot(data=dfc, x='rate')
plt.title('Ratings Boxplot')
plt.show()

# Cost (if present)
if 'approx_cost(for two people)' in dfc.columns:
    sns.histplot(data=dfc, x='approx_cost(for two people)', bins=30, kde=True)
    plt.title('Distribution of Approx. Cost (for two people)')
    plt.show()

    sns.boxplot(data=dfc, x='approx_cost(for two people)')
    plt.title('Cost Boxplot')
    plt.show()

# Votes (if present)
if 'votes' in dfc.columns:
    sns.histplot(data=dfc, x='votes', bins=30, kde=True)
    plt.title('Distribution of Votes')
    plt.show()
```

```python
# 4) Categorical: locations and cuisines
# Top locations by restaurant count
if 'location' in dfc.columns:
    top_locs = dfc['location'].value_counts().head(15).sort_values()
    sns.barplot(y=top_locs.index, x=top_locs.values, orient='h')
    plt.title('Top 15 Locations by Restaurant Count')
    plt.xlabel('Count'); plt.ylabel('Location')
    plt.show()

# Top cuisines by count
if 'cuisines' in dfc.columns:
    cuisines_expl = (dfc['cuisines'].dropna()
                     .str.split(',')
                     .explode()
                     .str.strip())
    top_cuis = cuisines_expl.value_counts().head(15).sort_values()
    sns.barplot(y=top_cuis.index, x=top_cuis.values, orient='h')
    plt.title('Top 15 Cuisines')
    plt.xlabel('Count'); plt.ylabel('Cuisine')
    plt.show()
```

```python
# 5) Price bands and comparisons
if 'approx_cost(for two people)' in dfc.columns:
    dfc['price_band'] = pd.cut(
        dfc['approx_cost(for two people)'],
        bins=[-np.inf, 500, 1500, np.inf],
        labels=['Budget', 'Medium', 'High']
    )
    sns.countplot(data=dfc, x='price_band', order=['Budget', 'Medium', 'High'])
    plt.title('Restaurant Count by Price Band')
    plt.show()

    sns.boxplot(data=dfc, x='price_band', y='rate', order=['Budget', 'Medium', 'High'])
    plt.title('Rating by Price Band')
    plt.show()
```

```python
# 6) Bivariate: ratings vs. cost / votes; categorical impacts
# Ratings vs cost
if {'rate', 'approx_cost(for two people)'}.issubset(dfc.columns):
    temp = dfc[['rate', 'approx_cost(for two people)']].dropna()
    sns.regplot(data=temp, x='approx_cost(for two people)', y='rate',
                scatter_kws={'alpha': 0.3, 's': 20}, line_kws={'color': 'red'})
    plt.title('Rating vs Approx. Cost (for two people)')
    plt.show()

# Ratings vs votes
if {'rate', 'votes'}.issubset(dfc.columns):
    temp = dfc[['rate', 'votes']].dropna()
    sns.regplot(data=temp, x='votes', y='rate',
                scatter_kws={'alpha': 0.3, 's': 20}, line_kws={'color': 'red'})
    plt.title('Rating vs Votes')
    plt.show()

# Online order / table booking vs rating
for col in ['online_order', 'book_table', 'listed_in(type)']:
    if col in dfc.columns:
        sns.boxplot(data=dfc, x=col, y='rate')
        plt.title(f'Rating by {col}')
        plt.show()
```

```python
# 7) Grouped insights: cuisines and locations with strong ratings
# Cuisines with high avg rating (min count threshold)
if 'cuisines' in dfc.columns:
    c_expl = (dfc[['rate', 'cuisines']]
              .dropna(subset=['rate', 'cuisines'])
              .assign(cuisine=lambda x: x['cuisines'].str.split(','))
              .explode('cuisine'))
    c_expl['cuisine'] = c_expl['cuisine'].str.strip()
    g = (c_expl.groupby('cuisine')['rate']
         .agg(['count', 'mean'])
         .query('count >= 100')
         .sort_values('mean', ascending=False)
         .head(15))
    print("Top cuisines by average rating (count >= 100):\n", g)

    sns.barplot(data=g.reset_index(), y='cuisine', x='mean', orient='h')
    plt.title('Top Cuisines by Avg Rating (min 100)')
    plt.xlabel('Average Rating'); plt.ylabel('Cuisine')
    plt.show()

# Locations with strong ratings (min count threshold)
if 'location' in dfc.columns:
    locg = (dfc.dropna(subset=['rate', 'location'])
            .groupby('location')['rate']
            .agg(['count', 'mean'])
            .query('count >= 100')
            .sort_values('mean', ascending=False)
            .head(15))
    print("Top locations by average rating (count >= 100):\n", locg)

    sns.barplot(data=locg.reset_index(), y='location', x='mean', orient='h')
    plt.title('Top Locations by Avg Rating (min 100)')
    plt.xlabel('Average Rating'); plt.ylabel('Location')
    plt.show()
```

```python
# 8) Correlation heatmap (numeric only)
num_cols = dfc.select_dtypes(include=[np.number]).columns
if len(num_cols) >= 2:
    corr = dfc[num_cols].corr(numeric_only=True)
    sns.heatmap(corr, annot=True, fmt=".2f", cmap="Blues", square=True)
    plt.title('Correlation Heatmap (Numeric Features)')
    plt.show()
```

```python
# 9) Text signals (reviews): simple frequent words (quick & optional)
if 'reviews_list' in dfc.columns:
    import re
    from collections import Counter

    texts = ' '.join(map(str, dfc['reviews_list'].dropna().sample(min(500, dfc['reviews_list'].dropna().shape[0]), random_state=42)))
    tokens = re.findall(r'[a-zA-Z]{3,}', texts.lower())
    stop = set("""the and for you are with this that was have had but not all out has too very from they were been
                  good great nice awesome amazing bad worst best tasty delicious quality service staff ambiance
                  place food taste price""".split())
    words = [w for w in tokens if w not in stop]
    top_words = Counter(words).most_common(25)
    print("Top review tokens:", top_words)

    # Plot
    if top_words:
        labels, vals = zip(*top_words)
        sns.barplot(x=list(vals), y=list(labels), orient='h')
        plt.title('Top Tokens in Reviews (sample)')
        plt.xlabel('Count'); plt.ylabel('Token')
        plt.show()
```

```python
# 10) Quick, actionable summaries
out = {}

if {'rate', 'approx_cost(for two people)'}.issubset(dfc.columns):
    out['price_band_stats'] = (dfc
        .dropna(subset=['rate', 'approx_cost(for two people)'])
        .assign(price_band=pd.cut(dfc['approx_cost(for two people)'], bins=[-np.inf, 500, 1500, np.inf],
                                  labels=['Budget', 'Medium', 'High']))
        .groupby('price_band')['rate']
        .agg(['count', 'mean'])
        .round(2)
    )

if 'votes' in dfc.columns and 'rate' in dfc.columns:
    out['highly_rated_popular'] = (dfc
        .dropna(subset=['rate', 'votes'])
        .query('rate >= 4.5')
        .nlargest(10, 'votes')[['name', 'location', 'rate', 'votes']]
    )

for k, v in out.items():
    print(f"\n{k}:\n{v}")
```
